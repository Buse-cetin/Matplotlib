```python
gruplar = ["A","B","C"]
veriler = [15,45,30]
plt.figure(figsize=(9,3))
plt.subplot(131)
plt.bar(gruplar,veriler)
plt.subplot(132)
plt.scatter(gruplar,veriler)
plt.subplot(133)
plt.plot(gruplar,veriler)
plt.suptitle("6 Mart Canl Yayn Denemeleri")
```
